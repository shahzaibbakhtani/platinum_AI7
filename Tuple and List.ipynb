### Lists basics

```python
lst = [55,77,88,99,66,45]
print(lst)
print(len(lst))
print(type(lst))
```

```python
lst[2] = 66
print(lst)
```

```python
string = "Shahzaib"               # slicing 
string[0:4]
```

```python
lst[0:3]            # will come from 0 - 2 (3-1)
```

```python
lst[-3:-1]           # you will start write a smaller number every time, whether slicing (positive or negative)
```

### List Methods 

###### All operations will be done by using () round brackets 

```python
list = [ 1,2,3,4]
list.append(5)                       # to add a number (it will add the new digit at last)
print(list)
```

```python
list = [5,4,3,2,1]
list.sort()                           # sorting the list (ascending order)
print(list)
```

```python
list.sort(reverse=True)
print(list)
```

```python
lst = ["a","b","c","d"]
```

```python
lst.reverse()
print(list)
```

```python
lst.insert(1,5)
print(lst)
```

```python
lst.remove("b")
print(lst)
```

```python
lst.pop(2)
print(lst)
```
