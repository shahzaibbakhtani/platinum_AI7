```python
class Student():
    def __init__(self, name, marks):
        self.name = name 
        self.marks = marks

    def avg_marks(self):
        total = sum(self.marks)
        print(f"Hi {self.name}, your avg marks are {round(total / len(self.marks), 2)}")
```

```python
class Account:
    def __init__(self,bal,acc):
        self.balance = bal
        self.Account_no = acc

    def credit(self,amount):
        self.balance += amount 
        print("PKR",amount,"was added to your account")
        print("total balance = ", self.get_balance())

    def debit(self, amount):
        self.balance -= amount 
        print("PKR",amount,"was deducted from your account")
        print("total balance = ", self.get_balance())

    def get_balance(self):
        return self.balance
    
```

```python
acc1 = Account(50000,43334)
```

```python
acc1.credit(5000)
acc1.debit(3000)
```

```python
acc1.debit(400000000)
```

```python
print(acc1.balance)               # this is an attribute, so no brackets are required 
print(acc1.get_balance())         # this is a method / function in class, so brackets are needed
```

### Inheritance 

```python
class Car():
    @staticmethod
    def start():
        print("Car Started...")

    @staticmethod
    def stop():
        print("Car stopped")
        
```

```python
class Honda(Car):
    def __init__(self,name,speed):
        self.name = name 
        self.speed = speed
        
```

```python
car1 = Honda("Civic", "100mph")
```

```python
print(car1.name)
car1.start()
car1.stop()
```

##### Multi Inheritance 

```python
class bike():
    @staticmethod
    def start():
        print("bike Started...")

    @staticmethod
    def stop():
        print("bike stopped")

class Suzuki(bike):
    def __init__(self,name,speed):
        self.name = name 
        self.speed = speed

class fast(Suzuki):
    def __init__(self,type):
        self.type = type

bike1 = fast("Electric")
```

```python
print(bike1.type)
bike1.start()
bike1.stop()
```

```python
bike2 = fast("Diesel")
bike2.start()
bike2.stop()
```

#### Multi Level Inheritance
###### The derived class can inherit multiple classes
###### Two parent and one child 

```python
class A :
    valueA = "Welcome to class A"

class B :
    valueB = "Welcome to class B"

class C (A,B):
    valueC = "Welcome to class C"

c1 = C()
```

```python
print(c1.valueA)
print(c1.valueB)
print(c1.valueC)
```

#### Super Method  
######  It is used to call the parent class from a child class. For methods.

```python
class Car:
    def __init__(self,type):
        self.type = type 
    @staticmethod
    def start():
        print("Car started...")
    @staticmethod
    def stop():
        print("Car is stopped")

class BYD(Car):
    def __init__(self,name,type):
        self.name = name 
        super().__init__(type)
```

```python
car1 = BYD("ABC", "Electric")
```

```python
car1.type
```

```python
car1.start()
```

```python
class Stu :
    def __init__(self, phy,chem,math):
        self.phy = phy 
        self.chem = chem 
        self.math = math 
    @property
    def percentage(self):
        return str((self.phy+self.chem+self.math)/3) + "%"           # we use return in this not self.percentage in property 
```

### Polymorphism 

```python
class Complex:
    def __init__(self,real,img):
        self.real = real 
        self.img = img 
    def show_num (self):
        print(self.real, "i +" ,self.img, "j")

    def __add__(self,num2):
        new_real = self.real + num2.real
        new_img = self.img +num2.img
        return Complex(new_real , new_img)
        
    def __sub__(self,num2):
        new_real = self.real - num2.real
        new_img = self.img - num2.img
        return Complex(new_real , new_img)
```

```python
num1 = Complex(3,4)
num1.show_num()
```

```python
num2 = Complex(5,6)
num2.show_num()
```

```python
num3 = num1 + num2
num3.show_num()
```

```python
num4 = num1 - num2
num4.show_num()
```
