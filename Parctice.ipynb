```python
customers = {
    "Customer_ID": [1, 2, 3, 4, 5, 6],
    "Payment_ID": [201, 202, 203, 204, 205,206],
    "Name": ["Ali", "Sara", "Ahmed", "Zara", "Bilal", "Hina"],
    "City": ["Karachi", "Lahore", "Islamabad", "Karachi", "Lahore", "Islamabad"]
}
```

```python
payments = {
    "Payment_ID": [201, 202, 203, 204, 205],
    "Order_ID": [101, 102, 104, 105, 108],
    "Payment_Method": ["Card", "Cash", "Card", "Cash", "Card"],
    "Status": ["Paid", "Paid", "Pending", "Paid", "Failed"]
}
```

```python
orders = {
    "Order_ID": [101, 102, 103, 104, 105, 106, 107],
    "Customer_ID": [1, 2, 2, 3, 5, 6, 7],
    "Product": ["Laptop", "Phone", "Tablet", "Shoes", "Watch", "Bag", "Camera"],
    "Amount": [120000, 80000, 60000, 7000, 15000, 5000, 90000]
}
```

```python
df1 = pd.DataFrame(customers)
df2 = pd.DataFrame(payments)
df3 = pd.DataFrame(orders)
```

```python
df1.merge(df3)     # inner join
```

```python
df1.merge(df3 ,  how = "outer" , on = "Customer_ID")
```

```python
df1.merge(df2 , how = "left" , on = "Payment_ID")
```

```python
df1.merge(df2, how = "cross")
```

```python
pd.concat([df1,df2] , join = "inner")
```

```python
df1.merge(df3 , how = "right" )
```
