### Concatenate and merge data frames (joins)

```python
data = {"Name" : ["Ali" , "Shahzaib" , "Asad", "Asif"] , "ID" : [1,2,3,4]}
```

```python
df1 = pd.DataFrame(data)
df1
```

```python
df2 = pd.DataFrame({"ID" : [1,2,2,3,4,5] , "Score" : [88,78,77,89,59,100]})
```

```python
pd.concat([df1,df2] , axis = 0)      # this only appended row wise / top on top
```

```python
pd.concat([df1,df2] , axis = 1)       # join through coulmns/ side on side 
```

### Merge (join)

```python
pd.merge(df1,df2, how = "inner" , on = "ID")
```

```python
pd.merge(df1,df2, how = "left" , on = "ID")
```

```python
pd.merge(df1,df2, how = "right" , on = "ID")   # here ID is in both tables, so we use simple on, if  both tables had diffrent names, then we can use left on and righton 
```
