```python
class Car:
    def __init__(self, mode):
        self.mode = mode
        print(f"Car mode is: {self.mode}")

class Brand(Car):
    def __init__(self, mode, name):
        # 1. Receive 'mode' and 'name'
        # 2. Pass 'mode' up to Car
        super().__init__(mode)
        self.name = name 

class Toyota(Brand): 
    def __init__(self, mode, name, car_type):
        # 1. Receive all 3 attributes
        # 2. Pass 'mode' and 'name' up to Brand
        super().__init__(mode, name)
        self.type = car_type
```

```python
c1 = Toyota("Fast","Civic","electric")
```

```python
c1.name
c1.mode
```
