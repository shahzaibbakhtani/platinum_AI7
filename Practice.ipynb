```python
import matplotlib.pyplot as plt 
```

```python
x = [1,2,3,4]
y = [10,15,20,25]
```

```python
plt.figure(figsize=(3,4))
plt.plot(x,y, color = "g" , linestyle ="--", marker ="o")
plt.title("Line Graph")
plt.xlabel("X")
plt.ylabel("Y")
```

```python
x = [1,2,3,4]
y1 = [10,15,20,25]
y2 = [100,200,300,400]

plt.plot(x,y1, label = "2025 Sales", color = "g")
plt.plot(x,y2, label = "2026 Sales", color = "r")
plt.title("Combined Sales")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.legend()
plt.show()
```

```python
plt.figure(figsize= (4,4))
x = [1,2,3,4]
y = [10,15,20,25]
plt.bar(x,y)
plt.title("WOW")
plt.show()
```

```python
import random 
import numpy as np 
```

```python
x = np.random.randint(1,10,10) 
y = np.random.randint(1,100,10)
plt.plot(x,y)
plt.title("Random digits")
plt.show()
```

```python
x = ["Mon" , "Tue", "Wed"]
y = [100,200,500]
plt.pie(y, labels = x , autopct = "%1.1f%%")
plt.title("Wow")
plt.show()
```

```python
x = ["Mon" , "Tue", "Wed", "Thur"]
y = [100,200,500, 600]
y1 = [1,2,3,4]

plt.figure(figsize = (10,4))

plt.subplot(1,3,1)
plt.bar(x,y,color = "g")
plt.title("Bar Graph")

plt.subplot(1,3,2)
plt.scatter(y,y1)
plt.title("Scatter Plots")

plt.subplot(1,3,3)
plt.pie(y,labels = x , autopct = "%1.1f%%")
plt.title("Pie Chart")

plt.show()
```
