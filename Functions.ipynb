##### Currency Converter

```python
def convert(dollar,pkr=280):                                           # Input
    pkr = dollar * pkr                                                 # what the function will do
    print(dollar, "USD =", pkr, "pkr")                                 # Output
    
```

```python
convert(19,282)                                                      # custom (when the rate is increased) 
```

```python
convert(10)
```

```python
def convert(d):
    p = d * 280
    return p

print("dollar 10 =", convert(10), "PKR")  # dollar 10 = 2800 PKR
```

### Recursion basics

```python
def show(n):
    if n == 0:          # Base case, just like stoping condition for loop 
        return(n)
    print(n)
    show(n-1)
```

```python
show(5)
```

```python
def fact(n):
    if n == 1 or n== 0:
        return 1 
    return(n-1)* n
```

```python
fact(3)
```

```python
def sum_2(n):
    if n == 0:
        return 0
    return sum_2(n - 1) + n  # ✅ recursive call
```

```python
sum_2(10)
```
