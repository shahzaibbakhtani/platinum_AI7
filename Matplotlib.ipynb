```python
import matplotlib.pyplot as plt 
```

```python
# first chart 
x = [1,2,3,4]
y = [10,20,30,40]

plt.plot(x,y)   # by deafult it is line chart 
plt.show()      # good practice --> to display chart
```

```python
# cutomise chart 
plt.plot(x,y, color = "violet" , marker = "o", linestyle = "dashed" , linewidth = 1, markersize = 9)
plt.show()
```

```python
plt.figure(figsize= (4,4))       # helps to maintain chart size , length. figsize value is in inches 
plt.plot(x,y, color = "y" , marker = "o", linestyle = "--" , linewidth = 1, markersize = 9)
plt.title("First Chart")
plt.xlabel("X axis ka label")
plt.ylabel("Y axis ka label")
plt.show()
```

#### Advanced Line chart 

```python
plt.figure(figsize = (4,5))
x = [1,2,3,4]
y1 = [5,10,15,20]
y2 = [10,20,30,40]

plt.plot(x,y1,label = "2025 sales data" )
plt.plot(x,y2,label = "2026 sales data" )

plt.title("Year on Year sales")
plt.xlabel("Months")
plt.ylabel("Sales")

plt.legend()     # --> this gives the box 
plt.show()
```

#### Most Used charts 

```python
# Bar Charts 
plt.figure(figsize =(3,4))
x = [1,2,3,4]
y = [5,100,44,20]
plt.bar(x,y)
plt.title("Bar Chart")
plt.show()
```

### Histogram 

```python
# used for data distribution 
import random 
data = [random.randint(1,13) for i in range(100)]
plt.hist(data, bins= 10 )   # bins -->   kitne bars bane ghe
plt.title("Histogram")
plt.show()
```

### Pie chart 
###### Use to show part to whole relationship 

```python
plt.figure(figsize= (3,3))
x = ['A', 'B', 'C', 'D']
y = [10,20,30,40]
plt.pie(y , labels =x, autopct = "%1.1f%%")
plt.title("Pie Chart") 
plt.show()
```

#### Scatter Plots 
###### Scatter plots are used to show the relationship between two variables.

```python
plt.figure(figsize= (3,3))
x1 = [10,20,30,40]
x2 = [5,10,15,20]
plt.scatter(x1,x2)
plt.title("Scatter Plot")
plt.show()
```

### Sub Plots 
###### Subplots let you show multiple charts in one figure

```python
# bar chart 
x = ['Mon', 'Tue', 'Wed', 'Thur']
y = [10,20,30,40]

# Data 2 --> scater plot 
x1 = [5, 10, 15, 20]
x2 = [10,20,30,40]

plt.figure(figsize = (10,4))
plt.subplot(1,2,1)  # row, column , position 
plt.bar(x,y)
plt.title("Daily Sales")
plt.xlabel("Week Days")
plt.ylabel("Sales")

plt.subplot(1,2,2)
plt.scatter(x1,x2)
plt.title("Days on days Sales")

plt.show()
```

### Matplotlib with pandas 

```python
import pandas as pd
```

```python
data = { "Salary": [
        95000, 72000, 105000, 88000, 62000, 
        115000, 54000, 91000, 78000, 83000
    ],
    "Position": [
        "Software Engineer", "Data Scientist", "Project Manager", "UX Designer", "Intern",
        "CTO", "Marketing Associate", "DevOps Engineer", "Sales Lead", "Product Owner"
    ]
}
```

```python
df = pd.DataFrame(data)
df
```

```python
plt.figure(figsize = (30,10))
plt.bar(df["Position"] , df["Salary"], color = "g")
plt.title("Salary by position")
plt.xlabel("Positions")
plt.ylabel("Salary")
plt.savefig("First save.png")
plt.show()
```
