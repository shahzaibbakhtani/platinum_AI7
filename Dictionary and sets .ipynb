# Dictionary and Sets 
## Dictionary

```python
marks = {"Shahzaib": 76, "Asad" : 80, "Arfa" : 88, "Asif" : 77}
```

```python
print(marks, type(marks))
```

```python
print(marks["Asad"])        # dic is great for lookups 
```

```python
print(marks.items())
print(marks.keys())
print(marks.values())
```

```python
marks.update({"Asad":56}) 
print(marks)                                    # through update we can do changes in inside the dictionary 
```

```python
marks.update({"Shireen" : 100, "Assar" : 89})
print(marks)                                        # we can also add new keys and values in existing dictionary 
```

```python
print(marks.get("Shahzaib"))
```

```python
print(marks.get("Bakhtani"))                #  print(marks["bakhtani"]) --> this will give error , but that code will give none as an answer
```

```python
add = marks.setdefault("Soomar", 98)         # add new value if not there 
print(add)
print(marks)
```

```python
marks.popitem()                               # remove the last added value and key 
print(marks)
```

```python
add = marks.setdefault("Shahzaib", 76)        # if it exists in dic it does nothing 
print(marks)
```

```python
print(marks.get("Soomar"))
```

```python
marks2 = marks.copy()
print(marks2)
```

```python
marks2.clear()                   # clear the dictionary 
print(marks2)
```

## Sets 

```python
Set = {1,3,44}
```

```python
set1 = set()
print(set1)
print(type(set1))
```

```python
Set.add(66)
print(Set, type(Set))
len(Set)
```

```python
s1 = {1 ,3,45,56}
s2 = {2 ,22,45, 3}
print(s1.union(s2))          # select common terms as one
```

```python
print(len(s1)+len(s2))
union = s1.union(s2)
print(len(union))
```

```python
print(s1.intersection(s2))       # select same only
```

```python
print({1,3}.issubset(s1))
```
